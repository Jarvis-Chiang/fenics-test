```python
import show_vector_with_pyvista as swp
import numpy as np
sw = swp.ShowWithPyvista('output/theta_vector(adam).txt')
print(sw)
sw.add_grid()
sw.add_vector_field(mag=0.25)
# sw.add_grid()
# 创建从 (0, 0, 0) 到 (0, 6, 0) 之间的起始点
# start_points = np.array([[0, y, 0] for y in np.linspace(0, 6, num=50)])
n_points = 100
start_points = np.zeros((n_points, 3))
start_points[:, 0] = np.random.uniform(0, 9, n_points)
start_points[:, 1] = np.random.uniform(0, 6, n_points)
start_points[:, 2] = np.random.uniform(0, 0, n_points)
# print(start_points)

sw.add_streamlines(start_points=start_points)
sw.show()
```

```python
# 为了进一步提高流线的均匀性，我们可以采取以下措施：

# 1. 使用更复杂的起始点分布
# 2. 增加起始点数量
# 3. 在边界附近添加额外的起始点
# 4. 进一步调整流线参数

import numpy as np

# 定义网格的范围
x_min, x_max = 0, 9
y_min, y_max = 0, 6
z_min, z_max = 0, 0

# 增加网格分辨率
nx, ny = 50, 40  # 进一步增加x和y方向上的点数

# 使用组合分布方法创建起始点
x = np.concatenate([
    np.linspace(x_min, x_max, nx//2),  # 线性分布
    np.random.uniform(x_min, x_max, nx//2)  # 随机均匀分布
])
y = np.concatenate([
    np.linspace(y_min, y_max, ny//2),
    np.random.uniform(y_min, y_max, ny//2)
])
xx, yy = np.meshgrid(x, y)
zz = np.zeros_like(xx)

start_points = np.column_stack((xx.ravel(), yy.ravel(), zz.ravel()))

# 在边界附近添加额外的起始点
border_points = 100
border_x = np.random.choice([x_min, x_max], border_points)
border_y = np.random.uniform(y_min, y_max, border_points)
border_z = np.zeros(border_points)

start_points = np.vstack((start_points, np.column_stack((border_x, border_y, border_z))))

# 调整流线参数
integration_direction = 'both'
max_time = 20  # 增加最大积分时间
initial_step_length = 0.03  # 进一步减小初始步长，使流线更平滑
step_unit = 'cl'  # 使用特征长度作为步长单位

# 生成流线
sw.add_streamlines(start_points=start_points, 
                   integration_direction=integration_direction,
                   max_time=max_time,
                   initial_step_length=initial_step_length,
                   step_unit=step_unit)

# 调整视角和颜色
sw.plotter.view_xy()  # 设置为俯视图
sw.plotter.set_background('white')  # 设置背景为白色

# 添加颜色条以显示流线速度
sw.plotter.add_scalar_bar(title="速度", n_labels=5, position_x=0.8, position_y=0.1)

# 显示结果
sw.show()

# 如果流线仍然不够均匀，可以考虑以下方法：
# 1. 调整vector_field的mag参数（在之前的代码中）
# 2. 使用自适应的流线生成算法
# 3. 在特定区域增加更多的起始点
# 4. 尝试不同的积分器（例如，使用Runge-Kutta方法）
```
