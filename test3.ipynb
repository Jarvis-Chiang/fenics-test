```python
from dolfin import Mesh, XDMFFile, plot
import matplotlib.pyplot as plt

# 步骤1: 使用DOLFIN读取XML文件
xml_input = 'input/topo_shape.xml'
mesh = Mesh(xml_input)

print(f"网格信息:")
print(f"  - 维度: {mesh.geometric_dimension()}")
print(f"  - 节点数: {mesh.num_vertices()}")
print(f"  - 单元数: {mesh.num_cells()}")

# 步骤2: 显示网格
plt.figure(figsize=(10, 8))
plot(mesh)
plt.title("2D mesh")
plt.axis('equal')
plt.show()
```
