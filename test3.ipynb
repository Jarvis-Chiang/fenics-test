```python
from dolfin import *
import matplotlib.pyplot as plt
import numpy as np

# 读取 XML 网格文件
mesh = Mesh("/home/wsl-20/fenicsprojects/让我测试一下/input/banshou.xml")

# 提取顶点坐标
coordinates = mesh.coordinates()

# 提取单元
cells = mesh.cells()

# 创建一个图形
plt.figure()

# 绘制网格
for cell in cells:
    polygon = np.append(cell, cell[0])  # 完整的多边形
    plt.plot(coordinates[polygon, 0], coordinates[polygon, 1], 'k-')

# 设置坐标轴比例
plt.gca().set_aspect('equal')

# 显示图形
plt.show()
```
